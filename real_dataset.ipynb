# GENERATE METADATA

```python
import json
import os
```

```python
FEATURE_CATEGORICAL = 'categorical'
FEATURE_NUMERIC = 'numeric'

features = {
    'CLIENTNUM': FEATURE_NUMERIC,
    'Attrition_Flag': FEATURE_CATEGORICAL,
    'Customer_Age': FEATURE_NUMERIC,
    'Gender': FEATURE_CATEGORICAL,
    'Dependent_count': FEATURE_NUMERIC,
    'Education_Level': FEATURE_CATEGORICAL,
    'Marital_Status': FEATURE_CATEGORICAL,
    'Income_Category': FEATURE_CATEGORICAL,
    'Card_Category': FEATURE_CATEGORICAL,
    'Months_on_book': FEATURE_NUMERIC,
    'Total_Relationship_Count': FEATURE_NUMERIC,
    'Months_Inactive_12_mon': FEATURE_NUMERIC,
    'Contacts_Count_12_mon': FEATURE_NUMERIC,
    'Credit_Limit': FEATURE_NUMERIC,
    'Total_Revolving_Bal': FEATURE_NUMERIC,
    'Avg_Open_To_Buy': FEATURE_NUMERIC,
    'Total_Amt_Chng_Q4_Q1': FEATURE_NUMERIC,
    'Total_Trans_Amt': FEATURE_NUMERIC,
    'Total_Ct_Chng_Q4_Q1': FEATURE_NUMERIC,
    'Avg_Utilization_Ratio': FEATURE_NUMERIC,
    
}
```

```python
os.makedirs('./Datasets/Real/BankChurners/', exist_ok=True)
json.dump(features, open('./Datasets/Real/BankChurners/metadata.json', 'w'))
```

# GENERATE SEMI-SYNTHETIC DATASETS

```python
os.makedirs('./Datasets/Real/BankChurners/Clear/', exist_ok=True)

for i in range(1, 31):
    df.sample(3000, replace=True, random_state=i)\
        .to_csv(f'./Datasets/Real/BankChurners/Clear/dataset_{i}.csv', index=False)
```
